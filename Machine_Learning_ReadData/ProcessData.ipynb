**Yêu cầu 4.2:**
Thông kê dữ liệu missing cho từng thuộc tính, và liệt kê ra các bản ghi bị missing tương ứng với thuộc tính đó.

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY!}
def missing_data(df):
    miss_data = DataFrame.isnull().sum()
    miss_data_percent = 100 * DataFrame.isnull().sum() / len(df)
    miss_data_table = pd.concat([miss_data, miss_data_percent], axis=1)
    miss_data_table_ren_columns = miss_data_table.rename(
        columns = {0 : 'Số giá trị bị mất', 1 : 'Tỷ lệ % mất'})
    miss_data_table_ren_columns = miss_data_table_ren_columns[
        miss_data_table_ren_columns.iloc[:,1] != 0].sort_values(
        'Tỷ lệ % mất', ascending=False).round(2)
    print("File dữ liệu bao gồm: " + str(DataFrame.shape[1]) + 
         " cột.\n" "có " + str(miss_data_table_ren_columns.shape[0])
         + " cột chứa bị mất dữ liệu.")
    return miss_data_table_ren_columns
```
